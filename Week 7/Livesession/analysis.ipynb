```python
## Import the Required libraries
import pandas as pd
import numpy as np
```

```python
# Import the data
df = pd.read_csv('todays_data.csv') 

# Display the first few rows of the dataframe
#print(df)

# View the first 5 rows
df.head()
```

```python
# Describe the dataframe to get an overview of the data
df.describe().T
```

```python
# Get the unique count of values in the gender column
df['Gender'].value_counts()
```

```python
# Academic level of Learners
df['Academic Level'].value_counts()
```

```python
# Plot a line grapgh of the academic level of learners
df['Academic Level'].value_counts().plot(kind='line', 
                                         title='Academic Level of Learners', 
                                         xlabel='Academic Level', 
                                         ylabel='Count of Learners')
```

```python
# plot a bar graph of the academic level of learners
df['Academic Level'].value_counts().plot(kind='bar', 
                                         title='Academic Level of Learners', 
                                         xlabel='Academic Level', 
                                         ylabel='Count of Learners')
```

```python
# Use a pie chart to show the academic level of learners
df['Academic Level'].value_counts().plot(kind='pie', 
                                         title='Academic Level of Learners', 
                                         xlabel='Academic Level', 
                                         ylabel='Count of Learners', 
                                         autopct='%1.1f%%') # Show percentage on the pie chart
```
