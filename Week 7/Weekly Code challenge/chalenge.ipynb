# Python Data Analysis Challenge

```python
# import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a dictionary with the data
data = {
    'Date': ['2025-03-01', '2025-03-01', '2025-03-02', '2025-03-02'],
    'Product': ['Laptop', 'Mouse', 'Laptop', 'Keyboard'],
    'Quantity Sold': [5, 15, 3, 8],
    'Revenue ($)': [5000, 300, 3000, 800]
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('sales_data.csv', index=False)
```

## Load the CSV file

```python
sales_data = pd.read_csv('sales_data.csv')

# Print the first few rows to verify the data
print(sales_data.head())
```

## Calculate Total Revenue

```python
total_revenue = sales_data['Revenue ($)'].sum()
print(f'Total Revenue: ${total_revenue}')
```

## Find Best-Selling Product

To find the best-selling product, we need to group the data by product and calculate the total quantity sold for each product. We can then find the product with the highest quantity sold.

```python
# Group the data by product and calculate total quantity sold
product_sales = sales_data.groupby('Product')['Quantity Sold'].sum().reset_index()

# Find the best-selling product
best_selling_product = product_sales.loc[product_sales['Quantity Sold'].idxmax()]
print(f'Best-Selling Product: {best_selling_product["Product"]} ({best_selling_product["Quantity Sold"]} units sold)')
```

## Identify Day with Highest Sales

To identify the day with the highest sales, we can group the data by date and calculate the total revenue for each day.

```python
# Group the data by date and calculate total revenue
date_revenue = sales_data.groupby('Date')['Revenue ($)'].sum().reset_index()

# Find the day with the highest sales
highest_sales_day = date_revenue.loc[date_revenue['Revenue ($)'].idxmax()]
print(f'Highest Sales Day: {highest_sales_day["Date"]}')
```

## Save Results to File

```python
# Save the results to a file
with open('sales_summary.txt', 'w') as f:
    f.write(f'Total Revenue: ${total_revenue}\n')
    f.write(f'Best-Selling Product: {best_selling_product["Product"]} ({best_selling_product["Quantity Sold"]} units sold)\n')
    f.write(f'Highest Sales Day: {highest_sales_day["Date"]}\n')
```

## Print the insights

```python
print(f'Total Revenue: ${total_revenue}')
print(f'Best-Selling Product: {best_selling_product["Product"]} ({best_selling_product["Quantity Sold"]} units sold)')
print(f'Highest Sales Day: {highest_sales_day["Date"]}')
```

## Bonus Challenge: Visualize Sales Trends

```python
# Group the data by date and calculate total revenue
date_revenue = sales_data.groupby('Date')['Revenue ($)'].sum().reset_index()

# Plot the sales trends
plt.plot(date_revenue['Date'], date_revenue['Revenue ($)'])
plt.xlabel('Date')
plt.ylabel('Revenue ($)')
plt.title('Sales Trends')
plt.show();
```

On March 1, the revenue was higher at approximately $5200, while on March 2, it dropped significantly to around $3800. This decline in revenue could indicate lower sales activity or fewer high-value products sold on the second day.
