# Data Visualization with Matplotlib

Matplotlib is used to create charts and graphs like line graphs, bar charts, pie charts, and more.

## üß∞¬†What is Matplotlib?

Matplotlib is a library that allows you to create visual representations of your data. It‚Äôs especially useful when working with data in Pandas or NumPy.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

## üìà Basic Line Plot

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

# Create a line plot
plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show();
```

## üìä Bar Chart Example

```python
# Sample data
names = ['Alice', 'Bob', 'Charlie']
scores = [85, 90, 78]

plt.bar(names, scores, color='skyblue')
plt.title("Student Scores")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show();
```

## üç∞ Pie Chart Example

```python
# Sample data
activities = ['Sleeping', 'Eating', 'Coding', 'Gaming']
hours = [8, 2, 8, 6]

plt.pie(hours, labels=activities, autopct='%1.1f%%')
plt.title("Daily Activities")
plt.show();
```

## üìè Histogram Example

```python
# Example: showing frequency of values
ages = [22, 21, 20, 23, 24, 22, 20, 21, 22, 25, 23]

plt.hist(ages, bins=5, color='purple')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show();
```

## üí° Combine with Pandas Example

```python
# Create a DataFrame
data = {
    'Year': [2021, 2022, 2023],
    'Users': [1500, 3000, 5000]
}

df = pd.DataFrame(data)

# Plot using pandas + matplotlib
plt.plot(df['Year'], df['Users'], marker='o')
plt.title("User Growth Over Time")
plt.xlabel("Year")
plt.ylabel("Users")
plt.grid(True)
plt.show();
```

## Solution 1: Countries Population Bar Chart

```python
# Sample data for countries and their populations (in millions)
countries = ['China', 'India', 'USA', 'Indonesia', 'Brazil']
populations = [1412, 1408, 332, 275, 215] # Population in millions

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(countries, populations, color='teal')
plt.title("Population by Country (in millions)")
plt.xlabel("Country")
plt.ylabel("Population (millions)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
#plt.savefig("population_by_country.png", dpi=300, bbox_inches='tight')
plt.show();
```

## Solution 2: Student's Day Pie Chart

```python
# Data for how a student spends 24 hours
activities = ['Sleep', 'Study', 'Classes', 'Eating', 'Exercise', 'Social/Leisure', 'Other']
hours = [8, 4, 5, 2, 1, 3, 1]

# Create a pie chart with percentages
plt.figure(figsize=(8, 8))
plt.pie(hours, labels=activities, autopct='%1.1f%%', startangle=90, 
        colors=['lightblue', 'lightgreen', 'coral', 'yellow', 'pink', 'skyblue', 'lightgray'])
plt.title("How a Student Spends 24 Hours")
plt.legend(activities, loc="center left", bbox_to_anchor=(1, 0.5))
#plt.savefig('line_chart.png')  # Save chart
plt.show();
```

## Solution 3: Daily Temperature Line Chart üìâ

```python
# Data for temperature changes during the day
times = ['Morning (6AM)', 'Noon (12PM)', 'Evening (6PM)', 'Night (12AM)']
temperatures = [15, 25, 22, 14]  # in Celsius

# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(times, temperatures, marker='o', linestyle='-', color='blue', linewidth=2, markersize=8)
plt.title("Temperature Changes Throughout the Day")
plt.xlabel("Time of Day")
plt.ylabel("Temperature (¬∞C)")
plt.grid(True, linestyle='--', alpha=0.7)
#plt.savefig('line_chart.png')  # Save chart
plt.show();
```
